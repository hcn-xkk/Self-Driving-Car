## Part 1: Load images and measurements

#### Helper functions for loading images:


```python
import pandas as pd
import cv2
import numpy as np
import random
import sklearn
```

```python
# Function to load center, left, right camera image names
def GetData( csv_dirs ):
    imagepaths = []
    measurements = []
    
    for csv_dir in csv_dirs:
        pds = pd.read_csv(csv_dir);
        pds_contents = pds.values.tolist()

        for line in pds_contents:
            line0 = '/'.join(line[0].split('\\')[-2:]).replace(" ", "")
            image_path = '/'.join(csv_dir.split('/')[:-1]) + '/' + line0
            steering = float(line[3])
            imagepaths.append(image_path)
            measurements.append(steering)

            line1 = '/'.join(line[1].split('\\')[-2:]).replace(" ", "")
            image_path = '/'.join(csv_dir.split('/')[:-1]) + '/' + line1
            steering = float(line[3]) + 0.2
            imagepaths.append(image_path)
            measurements.append(steering)

            line2 = '/'.join(line[2].split('\\')[-2:]).replace(" ", "")
            image_path = '/'.join(csv_dir.split('/')[:-1]) + '/' + line2
            steering = float(line[3]) - 0.2
            imagepaths.append(image_path)
            measurements.append(steering)
        
    return (imagepaths, measurements)
```

```python
# Generator function to prepare image data and steering data. Perform data augmentation by flipping images
def generator(samples, batch_size=32):
    # samples should have the form : list(zip( imagepaths, measurements ))
    num_samples = len(samples)
    while 1: # Loop forever so the generator never terminates
        random.shuffle(samples)

        for offset in range(0, num_samples, batch_size):
            batch_samples = samples[offset:offset+batch_size]
            imagepaths, measurements = zip(*batch_samples)
            images = []
            for image_path in imagepaths:
                image = cv2.imread(image_path)
                images.append(image)

            # Code for data augmentation: flip images
            flipped_images = []
            flipped_measurements = []
            for image, measurement in zip(images, measurements):
                flipped_images.append( np.fliplr(image) )
                flipped_measurements.append( -1.0*measurement ) 

            # group into arrays and output
            X_train = np.array(list(images) + flipped_images)
            y_train = np.array(list(measurements) + flipped_measurements)

            yield sklearn.utils.shuffle(X_train, y_train)
```

```python

batch_size = 256


samples1 = GetData( [ \
                    "/newdata/newdata/driving_log.csv", \
                    "/newdata/newdata2/driving_log.csv", \
                    "/newdata/mydata3/driving_log.csv", \
                    "/newdata/mydata6/driving_log.csv", \
                    "/newdata/mydata7/driving_log.csv", \
                    "/newdata/mydata8/driving_log.csv", \
                    "/newdata/mydata9/driving_log.csv", \
                    "/newdata/mydata16/driving_log.csv", \
                    ] )
samples2 = GetData( [ \
                    "/newdata/mydata13/driving_log.csv", \
                     ] )

from sklearn.model_selection import train_test_split
# Can separate training and validation using train_test_split
# train_samples, validation_samples = train_test_split(list(zip(samples[0], samples[1])), test_size=0.1)

# Here I use a specific trial for validation.
train_samples = list(zip(samples1[0], samples1[1]))
validation_samples = list(zip(samples2[0], samples2[1]))

train_generator = generator(train_samples, batch_size=batch_size)
validation_generator = generator(validation_samples, batch_size=batch_size)
```

## Part 2: Build CNN network for predicting steering angle


### Part 2.1 Apply LeNet architecture

#### Build LeNet model:


```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Lambda, Conv2D, MaxPooling2D, Activation, Cropping2D, Dropout

model = Sequential(
    [
        Cropping2D( cropping=((70, 25), (0, 0)) , input_shape = ( 160, 320, 3)),
        Lambda(lambda x: x/255.0 - 0.5),
        Conv2D(6, (5,5), padding="valid", activation='relu'),
        MaxPooling2D(pool_size = (2,2), padding="valid"),
        Conv2D(16, (5,5), padding="valid", activation='relu'),
        MaxPooling2D(pool_size = (2,2), padding="valid"),
        Dropout(0.5),
        Flatten(),
        Dense(120),
        Activation('relu'),
        Dense(84),
        Dense(1)
    ]
)

model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

#### Train LeNet model and save trained model:

model.fit_generator works well if there is large dataset and takes too much memory loading all the data.

model.fit works if for small dataset. 


```python
import math
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint('modelLeNet_{epoch:03d}_{loss:04f}_{val_loss:04f}.h5', verbose=1, monitor='val_loss') 

model.fit_generator(train_generator, \
            steps_per_epoch=math.ceil(len(train_samples)/batch_size), \
            validation_data=validation_generator, \
            validation_steps=math.ceil(len(validation_samples)/batch_size), \
            epochs=2, verbose=1, \
            callbacks=[checkpoint])
```

### Part 2.2: Apply Nvidia architecture

#### Build Nvidia model:

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Lambda, Conv2D, MaxPooling2D, Activation, Cropping2D
from tensorflow.keras import regularizers

model = Sequential(
    [
        Cropping2D( cropping=((70, 22), (0, 0)), input_shape = ( 160, 320, 3)),
        Lambda(lambda x: x/255.0 - 0.5),
        Conv2D(24, (5,5), padding="valid", activation='relu', kernel_regularizer=regularizers.l2(0.0005)),
        MaxPooling2D(pool_size = (2,2), padding="valid"),
        Conv2D(36, (5,5), padding="valid", activation='relu', kernel_regularizer=regularizers.l2(0.0005)),
        MaxPooling2D(pool_size = (2,2), padding="valid"),
        Conv2D(48, (5,5), padding="valid", activation='relu', kernel_regularizer=regularizers.l2(0.0005)),
        MaxPooling2D(pool_size = (2,2), padding="valid"),
        Conv2D(64, (3,3), padding="valid", activation='relu', kernel_regularizer=regularizers.l2(0.0005)),
        Conv2D(64, (3,3), padding="valid", activation='relu', kernel_regularizer=regularizers.l2(0.0005)),
        Flatten(),
        Dense(100, kernel_regularizer=regularizers.l2(0.0005)),
        Activation('relu'),
        Dense(50, kernel_regularizer=regularizers.l2(0.0005)),
        Activation('relu'),
        Dense(10, kernel_regularizer=regularizers.l2(0.0005)),
        Activation('relu'),
        Dense(1)
    ]
)

model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

#### Train Nvidia model and save trained model:


```python
import math
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint('/content/drive/My Drive/Colab_Notebooks/Udacity/Trial7/modelNvidiaNoDrop_{epoch:03d}_{loss:03f}_{val_loss:03f}.h5', \
                             verbose=1, monitor='val_loss') 

model_history = model.fit_generator(train_generator, \
            steps_per_epoch=math.ceil(len(train_samples)/batch_size), \
            validation_data=validation_generator, \
            validation_steps=math.ceil(len(validation_samples)/batch_size), \
            epochs=20, verbose=1, \
            callbacks=[checkpoint])
```

```python
# Plot loss history

plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.legend(['Loss','Val_loss'])
```

Comment: Several possible reasons for validation loss lower than training loss, 

(1) regularization loss is only applied to training but not validation. 

(2) trace used for validation is very good and is driving the center lane.

```python
model.evaluate(validation_generator, verbose=1, steps=math.ceil(len(validation_samples)/batch_size))
```
